```python
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import pandas as pd
import numpy as np

path = 'D:/桌面/My_TTP/EmRL/save/data20000_8.pkl'
with open(path,'rb') as f:
    data_dict = pickle.load(f)
# print(data_dict.type())
# data_list = data_dict[0]
x = [i for i in range(len(data_dict))]
data = pd.DataFrame({'episode':np.array(x[0:200]),
                    'rewards':np.array(data_dict[10000:10200]),
                    })

data['episode'] = data['episode'].astype('float64')
data['rewards'] = data['rewards'].astype('float64')

plt.figure(dpi=100)
sns.set(style="whitegrid",font_scale=1.2)
g = sns.lmplot(x="episode",y="rewards",data=data,
                order=4,
                ci=100,)

g.fig.set_size_inches(10,8)
plt.show()
print(data)
```
